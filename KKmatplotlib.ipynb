# bars

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D','E'])
y=np.array([3,6,7,9,2])
plt.bar(x,y,color='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D','E'])
y=np.array([3,6,7,9,2])
plt.barh(x,y,color='y')
plt.show()
```

```python
# The bar() takes the keyword argument width to set the width of the bars:
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D','E'])
y=np.array([3,6,7,9,2])
plt.bar(x,y,color='y',width=.8)
plt.show() 
```

```python
# The barh() takes the keyword argument height to set the height of the bars:
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['A','B','C','D','E'])
y=np.array([3,6,7,9,2])
plt.barh(x,y,color='b',height=.8)
plt.show() 
```

```python
y=np.array([35,15,5,25,20])
mylabels=['apple','banana','grapes','dates','kaju']
myexplodes=[.1,0,0,.2,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplodes,shadow=True)
plt.show()
```

```python
y=np.array([35,15,5,25,20])
mylabels=['apple','banana','grapes','dates','kaju']
myexplodes=[.1,0,0,.2,0]
mycolors=['r','y','b','black','g']
plt.pie(y,labels=mylabels,colors=mycolors,startangle=90,explode=myexplodes,shadow=True)
plt.legend()
plt.show()
```

```python
y=np.array([35,15,5,25,20])
mylabels=['apple','banana','grapes','dates','kaju']
myexplodes=[.1,0,0,.2,0]
mycolors=['r','y','b','black','g']
plt.pie(y,labels=mylabels,colors=mycolors,startangle=90,explode=myexplodes,shadow=True)
plt.legend(title="Five fruits:")
plt.show()
```
